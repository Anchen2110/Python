```python
import random as rd
signs=['X','O']
field=[[11,12,13],[21,22,23],[31,32,33]]
win_exist=False
def clear_board():
    field=[[11,12,13],[21,22,23],[31,32,33]]

def board_output():
    for i in range(3):
        print("____________________")
        for j in range(3):
            print("|",field[i][j],"|" , end=" ")
        print("\n")
        

def userInput(user_sign):
    user_ans_row = input("Please, input row number of " + user_sign +" ")
    if user_ans_row.isdigit():
        if (int(user_ans_row)>0) and (int(user_ans_row)<10):
            user_ans_col = input("Please, input column number of " + user_sign  +" ")
            if user_ans_col.isdigit():
                if (int(user_ans_col)>0) and (int(user_ans_col)<10):
                    print("Cool input!")
                    user_row=int(user_ans_row)
                    user_col=int(user_ans_col)
                    if field[user_row-1][user_col-1] not in signs:
                        print("ok!")
                        field[user_row-1][user_col-1]=user_sign
                    else:
                        print("But position is not free!!!")
                        userInput(user_sign)
                else:
                    print("Your column number is not range[1,9]")
                    userInput(user_sign)
                    
            else: 
                print("You column number isn't number")
                userInput(user_sign)
        else:
            print("Your row number is not range[1,9]")
            userInput(user_sign)
    else: 
        print("You row number isn't number")
        userInput(user_sign)

        
win_list=[["X","X","X"],["O","O","O"]]

def win_check():
    win=False
    for i in range(3):
        if field[i] in win_list:
            win=True
            break;
    
    if (field[0][0]==field[1][1]) and (field[1][1]==field[2][2]):
        win=True
    if (field[0][2]==field[1][1]) and (field[1][1]==field[2][0]):
        win=True
    
    
    if (field[0][0]==field[1][0]) and (field[1][0]==field[2][0]):
        win=True
    if (field[0][1]==field[1][1]) and (field[1][1]==field[2][1]):
        win=True 
    if (field[0][2]==field[1][2]) and (field[1][2]==field[2][2]):
        win=True    
    return win   

def random_user():
    user_list=["X","O"]
    return user_list[rd.randrange(0,len(user_list))]

def my_game():
    print("Let's generate 1st user")
    user1=random_user()
    print("1st user is "+user1+"-user")
    if  user1=="X":
        x_steps=(0,2,4,6,8)
        o_steps=(1,3,5,7)
    else:
        o_steps=(0,2,4,6,8)
        x_steps=(1,3,5,7)
    for step_counter in range(9):
        print ("step # "+ str(step_counter+1))
        if step_counter in x_steps:
            userInput('X')
            board_output()
            if (win_check()):
                win_exist=True
                print ("X-user, you are the winner")
                break
        else: 
            userInput('O')
            board_output()
            if (win_check()):
                win_exist=True
                print ("O-user, you are the winner")
                break
    if step_counter==8 and  win_exist==False:
        print ("Nobody is winner")
        
```
